# File 2: train_ai.py

This script remains simple. It takes the physics boundaries you defined above and runs thousands of simulated days so the AI can learn how to save money.

```python
from stable_baselines3 import PPO
from microgrid_env import MicroGridEnv
import time

print("1. Loading the MicroGrid Physics Engine...")
env = MicroGridEnv()

print("2. Initializing the PyTorch Neural Network (PPO)...")
model = PPO("MlpPolicy", env, verbose=1, learning_rate=0.0003)

print("3. Commencing AI Training (150,000 steps)...")
start_time = time.time()
model.learn(total_timesteps=150000)
end_time = time.time()

print(f"Training Complete! Time elapsed: {round(end_time - start_time, 1)} seconds.")

print("4. Saving the Brain...")
model.save("ppo_smart_microgrid")
print("Success. Model saved as 'ppo_smart_microgrid.zip'")
```

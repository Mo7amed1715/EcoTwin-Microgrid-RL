# **Test the AI**

```python
from stable_baselines3 import PPO
import numpy as np

# Load the brain
model = PPO.load("ppo_smart_microgrid")

# Fake Data: Battery dying (20%), no sun, high demand (80%), 10 PM.
fake_scenario = np.array([0.20, 0.0, 0.0, 0.0, 0.0, 0.80, -0.5, 0.866], dtype=np.float32)

action, _ = model.predict(fake_scenario, deterministic=True)
ai_command = float(action[0])

if ai_command < -0.1: status = "DISCHARGE"
elif ai_command > 0.1: status = "CHARGE"
else: status = "IDLE"

print(f"The RL math output was: {round(ai_command, 3)}")
print(f"The final hardware command is: {status}")
```
